```python
df.groupby(['Sex'])
df.head(10)
at=sns.barplot(y='Age',x='Sex',data=df)
for bars in at.containers:
    at.bar_label(bars)
# The bar labels display the mean age for each sex group.
```

```python
ax=sns.barplot(y='BMI',x='HighBP',data=df,palette='mako')
for bars in ax.containers:
    ax.bar_label(bars)
# The bar plot compares the average BMI for individuals grouped by their HighBP status.
```

```python
ax=sns.barplot(y='Diabetes_binary',x='HighChol',data=df,palette='Set2')
for bars in ax.containers:
    ax.bar_label(bars)
 #It compare the average likelihood of diabetes for individuals grouped by their HighChol status.
```

```python
sns.lineplot(x='Age',y='HeartDiseaseorAttack', data=df)
plt.show()
#display a line graph showing how the variable HeartDiseaseorAttack changes with Age
```

```python
ax=sns.barplot(y='GenHlth',x='HvyAlcoholConsump',data=df,palette='Set2')
plt.show()
for bars in ax.containers:
    ax.bar_label(bars)
#It shows how GenHlth values are distributed across categories of HvyAlcoholConsump.
    
```

```python
sns.histplot(data=df, x='Age',hue='Sex', bins=20)
plt.title('Age Distribution')
plt.show()
#It creates a histogram of the Age distribution from df, with separate color-coded distributions for each category in Sex. 
```

```python
sns.boxplot(x='Diabetes_binary', y='Age', data=df, palette='Set3')
plt.title('Age Distribution by Diabetes Status')
plt.xlabel('Diabetes (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
#This code generates a box plot to visualize the distribution of Age for people grouped by their Diabetes_binary status, using the data from the df DataFrame
```

```python
sns.countplot(x='Smoker', hue='Diabetes_binary', data=df, palette='coolwarm')
plt.title('Smoking Status by Diabetes')
plt.xlabel('Smoker (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.legend(title='Diabetes (0 = No, 1 = Yes)')
plt.show()
#This code generates a count plot to show the distribution of smoking status (Smoker) across categories of diabetes status (Diabetes_binary) using the data in df.
```

```python
sns.barplot(x='Diabetes_binary', y='PhysActivity', data=df, palette='Set1')
plt.title('Physical Activity and Diabetes')
plt.xlabel('Diabetes (0 = No, 1 = Yes)')
plt.ylabel('Average Physical Activity')
plt.show()
#This code creates a bar plot to compare the average physical activity (PhysActivity) for individuals with and without diabetes (Diabetes_binary) using the data in df.
```

```python
sns.countplot(x='NoDocbcCost', hue='Diabetes_binary', data=df, palette='husl')
plt.title('Healthcare Access by Diabetes Status')
plt.xlabel('Could Not See Doctor Due to Cost (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.legend(title='Diabetes (0 = No, 1 = Yes)')
plt.show()
#This code generates a count plot that visualizes the relationship between healthcare access (NoDocbcCost) and diabetes status (Diabetes_binary) using the data in df.
```

```python
sns.barplot(x='Diabetes_binary', y='MentHlth', data=df, palette='Blues')
plt.title('Mental Health by Diabetes Status')
plt.xlabel('Diabetes (0 = No, 1 = Yes)')
plt.ylabel('Average Mental Health Score')
plt.show()
#This code creates a bar plot to compare the average mental health score (MentHlth) for individuals with and without diabetes (Diabetes_binary) using the data in df
```

```python
sns.scatterplot(x='Age', y='BMI', hue='Diabetes_binary', data=df, palette='coolwarm', alpha=0.7)
plt.title('BMI vs Age by Diabetes Status')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(title='Diabetes (0 = No, 1 = Yes)')
plt.show()
#This code creates a scatter plot to visualize the relationship between Age and BMI with respect to Diabetes_binary status using the data in df
```
